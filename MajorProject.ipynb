```python

import cv2
import pytesseract
from PIL import Image
from docx import Document
from google.colab import files
from IPython.display import display, Image as IPImage

# Function to preprocess the image
def preprocess_image(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Denoise the image
    denoised = cv2.fastNlMeansDenoising(gray, None, 30, 7, 21)

    # Apply binary thresholding
    _, binary_image = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Save the preprocessed image for verification
    cv2.imwrite("preprocessed_image.jpg", binary_image)

    return binary_image

# Function to extract Telugu handwritten text
def telugu_handwritten_ocr(image_path):
    # Preprocess the image
    processed_image = preprocess_image(image_path)

    # Perform OCR using Tesseract with Telugu language
    extracted_text = pytesseract.image_to_string(processed_image, lang='tel', config='--psm 6')

    return extracted_text

# Function to save text to a .docx file
!pip install gtts
# Upload a Telugu handwritten image
print("Please upload a handwritten Telugu text image:")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Perform Telugu OCR
print("Extracting Telugu text from the image...")
extracted_text = telugu_handwritten_ocr(image_path)

# Display extracted text
print("Extracted Telugu Text:")
print(extracted_text)

text  = extracted_text
from gtts import gTTS
from google.colab import files

# Function to convert Telugu text to speech
def telugu_text_to_speech(text, output_filename="telugu_voice.mp3"):
    """
    Convert Telugu text to speech and save as an audio file.

    :param text: Telugu text to be converted to speech
    :param output_filename: Name of the output audio file (default: telugu_voice.mp3)
    """
    # Convert text to speech
    tts = gTTS(text=text, lang='te')  # 'te' is the language code for Telugu
    tts.save(output_filename)
    print(f"Audio saved as {output_filename}")
    return output_filename

# Example Telugu text (replace with your own text)
telugu_text = text

# Convert Telugu text to speech
print("Converting text to speech...")
audio_file = telugu_text_to_speech(telugu_text)

# Provide the audio file for download
print(f"Downloading the audio file: {audio_file}")
files.download(audio_file)
```

```python
!pip install gradio opencv-python-headless pytesseract gTTS Pillow
!apt-get install -y tesseract-ocr-tel
import gradio as gr
import cv2
import pytesseract
from PIL import Image
import numpy as np
from gtts import gTTS
import tempfile

# Preprocess image
def preprocess_image(image):
    image = image.convert("RGB")
    open_cv_image = np.array(image)
    open_cv_image = open_cv_image[:, :, ::-1].copy()

    gray = cv2.cvtColor(open_cv_image, cv2.COLOR_BGR2GRAY)
    denoised = cv2.fastNlMeansDenoising(gray, None, 30, 7, 21)
    _, binary = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return binary

# OCR + TTS pipeline
def ocr_and_tts(image):
    processed = preprocess_image(image)
    text = pytesseract.image_to_string(processed, lang='tel', config='--psm 6')

    if not text.strip():
        return "No text detected", None

    tts = gTTS(text=text, lang='te')
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        tts.save(fp.name)
        return text, fp.name

# Gradio Interface
demo = gr.Interface(
    fn=ocr_and_tts,
    inputs=gr.Image(type="pil"),
    outputs=["text", "audio"],
    title="📝 Telugu Handwritten OCR + Text-to-Speech",
    description="Upload an image with Telugu handwriting to extract text and listen to it as speech."
)

demo.launch()
```
